### `convert_data` (interactive)

Calls [`convert_data.py`](./convert_data.py) (same as `bash convert_data.sh …`). Prefer the CLI from the repo root for automation.

`analyze-thinking` needs **numpy**.

```python
import os
import subprocess
import sys
from pathlib import Path


def repo_root() -> Path:
    cwd = Path.cwd().resolve()
    if (cwd / "convert_data.py").is_file():
        return cwd.parent
    for d in (cwd, *cwd.parents):
        if (d / "LLaMA-Factory").is_dir() and (d / "scripts" / "convert_data.py").is_file():
            return d
    raise RuntimeError("Run from repo root, from scripts/, or inside a checkout that contains LLaMA-Factory.")


ROOT = repo_root()
os.environ["REPO_ROOT"] = str(ROOT)
PY = Path(sys.executable)
CLI = ROOT / "scripts" / "convert_data.py"


def run(*args: str) -> None:
    subprocess.run([PY, str(CLI), *args], check=True)


print("REPO_ROOT =", ROOT)
```
